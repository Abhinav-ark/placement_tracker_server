```python
import pandas as pd
```

```python
df_2022 = pd.read_excel('2022.xlsx')
df_2022
```

```python
df_2022.columns
```

```python
df_2022.drop(['ID', 'Start time', 'Completion time', 'Last modified time'], axis=1, inplace=True)
df_2022
```

```python
df_2022['Name'] = df_2022['Name'].str.replace(r"\[.*\]$", "")
df_2022.drop(['Full Name'], axis=1, inplace=True)
df_2022
```

```python
df_2022.drop(['Rollno'], axis=1, inplace=True)
df_2022
```

```python
# From email ID extract the content before @, upper case it and add it as rollNo
df_2022['rollNumber'] = df_2022['Email'].str.split('@').str[0].str.upper()
df_2022
```

```python
df_2022['Section'].value_counts()
```

```python
# Take last character alone from section
df_2022['Section'] = df_2022['Section'].str[-1]
df_2022
```

```python
df_2022['Company'].unique()
```

```python
df_2022['Company offer is'].unique()
```

```python
df_2022['Company offer is'].value_counts()
```

```python
df_2022[df_2022['Company offer is'] == 'Placement/Internship']['Role/Position'].unique()
```

```python
df_2022['Company offer is'].replace({'Placement/Internship': 'PPO'}, inplace=True)
df_2022
```

```python
df_2022['Company offer is'].value_counts()
```

```python
df_2022['Placement /Internship'].value_counts()
```

```python
df_2022['Placement /Internship'].replace({'CIR': '1', 'not thru CIR': '0'}, inplace=True)
df_2022['Placement /Internship'].value_counts()
```

```python
# Rename column
df_2022.rename(columns={'Placement /Internship': 'isOnCampus'}, inplace=True)
df_2022
```

```python
df_2022.rename(
    columns={
        'Email': "studentEmail",
        'Name': "studentName",
        'Section': 'studentSection',
        'Company': 'companyName',
        'Interview date and Time': 'placementDate',
        'Tentative Location': 'jobLocation',
        'Role/Position': 'jobRole',
        'Company CTC': 'ctc',
        'Company offer is': 'isIntern/isPPO',
        'rollNumber': 'studentRollNo'
    },
    inplace=True
)
df_2022
```

```python
# Split isIntern/isPPO into two columns isIntern and isPPO
df_2022['isIntern'] = df_2022['isIntern/isPPO']
df_2022['isPPO'] = df_2022['isIntern/isPPO']

df_2022.drop(['isIntern/isPPO'], axis=1, inplace=True)
df_2022
```

```python
df_2022['isIntern'].value_counts()
```

```python
df_2022['isIntern'].replace({'Placement': '0', 'Internship': '1', 'PPO': '0'}, inplace=True)
df_2022['isIntern'].value_counts()
```

```python
df_2022['isPPO'].value_counts()
```

```python
df_2022['isPPO'].replace({'Placement': '0', 'Internship': '0', 'PPO': '1'}, inplace=True)
df_2022['isPPO'].value_counts()
```
