## 2024 Batch

```python
df_2024 = pd.read_csv('../done/2024.csv')
df_2024
```

```python
## Add Students and get studentId

# studentRollNo,studentEmail,studentPassword,studentName,studentSection,studentGender,studentBatch,studentDept,isHigherStudies,isPlaced,

student_df = df_2024[['studentRollNo', 'studentEmail', 'studentPassword', 'studentName', 'studentSection', 'studentBatch', 'studentDept', 'isHigherStudies', 'isPlaced']]
student_df
```

```python
f = open('insertStudentData.sql', 'w')

for index, row in student_df.iterrows():
    f.write("insert into studentData (studentRollNo,studentEmail,studentPassword,studentName,studentSection,studentBatch,studentDept,isHigherStudies,isPlaced) values('" + row['studentRollNo'] + "','" + row['studentEmail'] + "','" + row['studentPassword'] + "','" + row['studentName'] + "','" + row['studentSection'] + "','" + str(row['studentBatch']) + "','" + row['studentDept'] + "'," + str(row['isHigherStudies']) + "," + str(row['isPlaced']) + ");\n")

f.close()
```

```python
## Add studentId to df_2024 which is index + 1

df_2024['studentId'] = df_2024.index + 1
df_2024
```

```python
# studentID,companyID,ctc,jobRole,placementDate,isIntern,isPPO,isOnCampus,isGirlsDrive

placement_df = df_2024[['studentId', 'companyId', 'ctc', 'jobRole', 'placementDate', 'isIntern', 'isPPO', 'isOnCampus', 'isGirlsDrive']]
placement_df.rename(columns={'companyId': 'companyID'}, inplace=True)
placement_df
```

```python
placement_df['ctc'] = placement_df['ctc'].astype(str)
```

```python
placement_df['jobRole'].fillna('NA', inplace=True)
```

```python
f = open('insertPlacementData.sql', 'w')

for index, row in placement_df.iterrows():
    f.write("insert into placementData (studentId,companyID,ctc,jobRole,placementDate,isIntern,isPPO,isOnCampus,isGirlsDrive) values(" + str(row['studentId']) + "," + str(row['companyID']) + "," + str(row['ctc']) + ",'" + row['jobRole'] + "','" + str(row['placementDate']) + "'," + str(row['isIntern']) + "," + str(row['isPPO']) + "," + str(row['isOnCampus']) + "," + str(row['isGirlsDrive']) + ");\n")

f.close()
```
